# __NÚMERO DE ESTUDANTES E MÉDIA DE IDADE SEPARADOS POR COR/RAÇA EM CADA UF BRASILEIRA__

Ultimamente a polarização politica que assola nosso planeta tem tomado pautas de conversa  tanto em lares familiares quanto ambientes de trabalho.

Uma das questões debatidas no Brasil, é a existência de cotas em universidades para negros, pardos, e mais algumas pessoas que de alguma forma foram prejudicadas pela sociedade.

Com base em dados obtidos do Censo da Educação Superior de 2018[1], esta análise demonstra o número de estudantes e a média de idade separando os resultados de acordo com sua cor/raça, em cada estado do país.

[1] http://download.inep.gov.br/microdados/microdados_educacao_superior_2018.zip


```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
low_memory=False

indice= [0,1,2,3,4]
regioes = ['Norte','Sul','CentroOeste','Nordeste','Sudeste']
regioescod = [[11,12,13,14,15,16,17],[41,42,43],[50,51,52,53],[21,22,23,24,25,26,27,28,29],[31,32,33,35]]
regioescod2 = [11,12,13,14,15,16,17,41,42,43,50,51,52,53,21,22,23,24,25,26,27,28,29,31,32,33,35]
estados = ['Rondonia','Acre','Amazonas','Roraima','Para','Amapa','Tocantins','Parana','Santa Catarina','Rio Grande do Sul','Mato Grosso do Sul','Mato Grosso','Goias','Distrito Federal','Maranhao','Piaui','Ceara','Rio Grande do Norte','Paraiba','Pernambuco','Alagoas','Sergipe','Bahia','Minas Gerais','Espirito Santo','Rio de Janeiro','Sao Paulo']
statename = random.sample(range(0,60), 60)
a=0
for  k in regioescod2:
    statename[k]=estados[a]
    a=a+1

df = pd.read_csv(r'C:\Users\Max\Documents\Fatec\3o sem\ED\EP4\microdados_ed_superior_2018\dados\DM_ALUNO.csv', delimiter='|', encoding='iso-8859-1', usecols=['CO_UF_NASCIMENTO', 'TP_COR_RACA','NU_IDADE'], dtype={"CO_UF_NASCIMENTO":str})

for i in indice:
    a = regioes[i]
    print('')
    print('')
    print('Região: ',a)
    l = regioescod[i]
    
    for j in l:
        a=str (j)
        nome = statename[j]
        cteste1 = df.groupby(['CO_UF_NASCIMENTO']).get_group(a)['TP_COR_RACA'].value_counts()
        cteste2 = cteste1.rename(({0:'Nao Declarou', 1: 'Branca', 2:'Preta',3:'Parda',4:'Amarela',5:'Indigena',9:'Indisponível'}))
        cteste0= df.groupby(['CO_UF_NASCIMENTO']).get_group(a).groupby(['TP_COR_RACA'])['NU_IDADE'].agg(np.mean)
        cteste3 =cteste0.rename(({0:'Nao Declarou', 1: 'Branca', 2:'Preta',3:'Parda',4:'Amarela',5:'Indigena',9:'Indisponível'}))
        print('Estado :', nome)
        print ('Codigo UF: ', a)
        fig, axs = plt.subplots(1,2, figsize = (20,10))
        cteste2.plot (ax=axs[0], kind = "bar",  grid = (True), x=("CO_UF_NASCIMENTO", "TP_COR_RACA"), y =('TP_COR_RACA'),  subplots = True, title = "Número de estudantes separados por cor/raça:                     ---                     Média de idade por cor/raça no Estado mencionado:")
        cteste3.plot ( ax=axs[1], kind = "bar",  grid = (True), x=("CO_UF_NASCIMENTO",  "TP_COR_RACA"), y =('NU_IDADE'),  subplots = True, color = ['blue', 'pink', 'black','brown','yellow','red','grey'])
        plt.show()
```

## Ao verificar estes gráficos, podemos verificar que a cor preta, no máximo chega a 3a colocação, também podemos notar que sua média de idade também é ligeiramente maior, o que indica um possivel ingresso tardio na universidade.
Ainda há uma grande disparidade entre as cores, mesmo com as "ajudas", fornecidas pelo governo brasileiro, isto sugere ao menos que as políticas atuais ainda não são totalmente eficazes.
